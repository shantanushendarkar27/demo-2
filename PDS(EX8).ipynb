Introduction

In our daily life, we often notice that some things are connected. For example, if a student studies more, their marks usually improve. Similarly, higher advertising may lead to higher sales. To study such relationships between variables, we use statistical tools like correlation and regression. These methods help us understand how variables are related and how one can be used to predict another.

1. What is Correlation? How is it Different from Regression?

Correlation is a method used to measure the relationship between two variables. It tells us whether the variables move together or in opposite directions.

If both variables increase or decrease together, it is called positive correlation.
If one increases while the other decreases, it is called negative correlation.
If there is no clear pattern, then there is no correlation.

The value of correlation lies between -1 and +1, where:

+1 means perfect positive relationship
-1 means perfect negative relationship
0 means no relationship

Regression, on the other hand, goes one step further. It not only shows the relationship but also helps us predict the value of one variable based on another.


Correlation tells us “how variables are related”
Regression tells us “how one variable affects another and helps predict it”
2. What is Simple Linear Regression? Explain with Example

Simple linear regression is a method used to describe the relationship between two variables using a straight line.

y=a+bx

In this equation:

y is the dependent variable (what we want to predict)
x is the independent variable (input)
a is the starting value (intercept)
b shows how much y changes when x changes (slope)
Example:

Suppose we want to predict a student’s marks based on the number of hours they study.

If we collect data and apply regression, we can draw a straight line that shows how marks increase with study time. Using this line, we can estimate marks for any given number of study hours.

3. How do you Calculate Pearson Correlation in Python?

Karl Pearson’s correlation coefficient is widely used to measure the strength of a linear relationship.

In Python, it can be calculated easily using libraries like NumPy.

import numpy as np

# Example data
x = [2, 4, 6, 8, 10]
y = [3, 5, 7, 9, 11]

# Calculate Pearson correlation
correlation = np.corrcoef(x, y)[0, 1]

print("Correlation:", correlation)

If the result is close to +1, it means a strong positive relationship. If it is close to 0, the relationship is weak.

4. What is Multicollinearity? Does it Affect Simple Regression?

Multicollinearity happens when two or more independent variables are strongly related to each other.

For example, if we include both “temperature in Celsius” and “temperature in Fahrenheit” in the same model, they will be highly correlated.

This creates confusion in determining which variable is actually influencing the result.

However, in simple linear regression, we only use one independent variable. So, multicollinearity is not a problem in simple regression. It mainly affects multiple regression models.

5. A Dataset Shows Strong Correlation but Poor Prediction Accuracy – Why?

Sometimes, even when correlation is strong, the predictions may not be accurate. This can happen due to several reasons:

The relationship may not be perfectly linear (it could be curved).
There may be extreme values (outliers) affecting the results.
Important variables may be missing from the model.
The dataset may be too small to give reliable predictions.
The model used may not be suitable for the data.

So, correlation alone is not enough for good prediction. We must also choose the right model and analyze the data carefully.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# --------------------------------------
# Step 1: Create Dataset
# --------------------------------------
# Example: Study Hours vs Marks
data = {
    "Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [2, 4, 5, 4, 5, 7, 8, 9]
}

df = pd.DataFrame(data)

print("Dataset:\n", df)

# --------------------------------------
# Step 2: Karl Pearson Correlation
# --------------------------------------
corr, p_value = pearsonr(df["Hours"], df["Marks"])

print("\nKarl Pearson Correlation Coefficient:", corr)

# --------------------------------------
# Step 3: Simple Linear Regression
# --------------------------------------
X = df[["Hours"]]   # Independent variable
y = df["Marks"]     # Dependent variable

model = LinearRegression()
model.fit(X, y)

# Get parameters
slope = model.coef_[0]
intercept = model.intercept_

print("\nRegression Equation:")
print(f"y = {intercept:.2f} + {slope:.2f}x")

# --------------------------------------
# Step 4: Prediction
# --------------------------------------
y_pred = model.predict(X)

# Predict for new value (e.g., 9 hours)
new_hours = np.array([[9]])
predicted_marks = model.predict(new_hours)

print("\nPredicted Marks for 9 hours study:", predicted_marks[0])

# --------------------------------------
# Step 5: Model Evaluation
# --------------------------------------
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("\nMean Squared Error:", mse)
print("R2 Score:", r2)

# --------------------------------------
# Step 6: Visualization
# --------------------------------------
plt.scatter(df["Hours"], df["Marks"], label="Actual Data")
plt.plot(df["Hours"], y_pred, label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Correlation and Regression Analysis")
plt.legend()
plt.show()
```
